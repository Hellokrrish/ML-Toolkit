### 3.What is the distribution of discounted prices vs. actual prices?


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('amazon.csv')

# Display the first few rows of the dataframe to understand its structure
print(df.head())

# Clean the price columns (assuming they contain currency symbols and commas)
df['actual_price'] = df['actual_price'].replace('[\₹,]', '', regex=True).astype(float)
df['discounted_price'] = df['discounted_price'].replace('[\₹,]', '', regex=True).astype(float)

# Plotting the distribution of Actual Prices vs. Discounted Prices
plt.figure(figsize=(14, 7))

# Plot Actual Prices
sns.histplot(df['actual_price'], kde=True, color='blue', label='actual_price', bins=30)

# Plot Discounted Prices
sns.histplot(df['discounted_price'], kde=True, color='orange', label='discounted_price', bins=30)

plt.title('Distribution of Actual Prices vs. Discounted Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```

### 4. How does the average discount percentage vary across categories?

```python
import pandas as pd

# Load your data into a pandas DataFrame
df = pd.read_csv('amazon.csv')

# Display the first few rows of the dataframe to understand its structure
print(df.head())

# Clean the price columns (remove currency symbols and commas)
df['actual_price'] = df['actual_price'].replace('[\₹,]', '', regex=True).astype(float)
df['discounted_price'] = df['discounted_price'].replace('[\₹,]', '', regex=True).astype(float)

# Clean the discount_percentage column
# Remove percentage symbols and convert to numeric
df['discount_percentage'] = df['discount_percentage'].replace('[\%]', '', regex=True).astype(float)

# Calculate the Discount Percentage if not already present or if it's invalid
if 'discount_percentage' not in df.columns or df['discount_percentage'].isnull().all():
    df['discount_percentage'] = ((df['actual_price'] - df['discounted_price']) / df['actual_price']) * 100

# Calculate the average discount percentage for each category
average_discount_percentage = df.groupby('category')['discount_percentage'].mean()

# Display the average discount percentage for each category
print(average_discount_percentage)

# Optional: Plotting the average discount percentage by category
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
average_discount_percentage.plot(kind='bar', color='skyblue')
plt.title('Average Discount Percentage by Category')
plt.xlabel('Category')
plt.ylabel('Average Discount Percentage')
plt.grid(axis='y')
plt.show()
```

### 6. What are the most popular product keywords?

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from collections import Counter
import matplotlib.pyplot as plt

# Load your data into a pandas DataFrame
df = pd.read_csv('amazon.csv')

# Display the first few rows of the dataframe to understand its structure
print(df.head())

# Combine 'product_name' and 'about_product' for keyword analysis
df['combined_text'] = df['product_name'].astype(str) + ' ' + df['about_product'].astype(str)

# Initialize CountVectorizer to extract keywords
vectorizer = CountVectorizer(stop_words='english', ngram_range=(1, 1))  # Unigrams (single words)
X = vectorizer.fit_transform(df['combined_text'])

# Convert the matrix to a DataFrame and sum the counts
keywords = vectorizer.get_feature_names_out()
counts = X.sum(axis=0).A1
keyword_counts = dict(zip(keywords, counts))

# Convert to a DataFrame for easier manipulation
keyword_counts_df = pd.DataFrame(list(keyword_counts.items()), columns=['Keyword', 'Count'])

# Sort by count in descending order
keyword_counts_df = keyword_counts_df.sort_values(by='Count', ascending=False)

# Display the top N keywords
top_n = 10  # Change this value as needed
top_keywords = keyword_counts_df.head(top_n)
print(top_keywords)

# Optional: Plot the most popular keywords
plt.figure(figsize=(12, 8))
plt.barh(top_keywords['Keyword'], top_keywords['Count'], color='skyblue')
plt.xlabel('Count')
plt.title('Most Popular Product Keywords')
plt.gca().invert_yaxis()
plt.show()
```

### 8. What is the correlation between discounted_price and rating?

```python
import pandas as pd

# Load your data into a pandas DataFrame
df = pd.read_csv('amazon.csv')

# Display the first few rows of the dataframe to understand its structure
print(df.head())

# Clean the discounted_price column (remove currency symbols and commas)
df['discounted_price'] = df['discounted_price'].replace('[\₹,]', '', regex=True).astype(float)

# Ensure the rating column is numeric
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

# Calculate the correlation between discounted_price and rating
correlation = df[['discounted_price', 'rating']].corr().iloc[0, 1]

# Display the correlation coefficient
print(f'Correlation between discounted_price and rating: {correlation:.4f}')
```

### 9.Does the duration of tracks vary significantly across different artists? Explore this visually using a box plot or violin plot.


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Select a subset of artists for clarity in the plot
selected_artists = df1['Artist'].value_counts().nlargest(10).index  # Selecting top 10 artists by track count
subset_data = df1[df1['Artist'].isin(selected_artists)]

# Create a box plot
plt.figure(figsize=(14, 8))
sns.boxplot(x='Artist', y='Duration (ms)', data=subset_data)
plt.title('Box Plot of Track Duration Across Artists')
plt.xticks(rotation=90)
plt.show()

# Create a violin plot
plt.figure(figsize=(14, 8))
sns.violinplot(x='Artist', y='Duration (ms)', data=subset_data)
plt.title('Violin Plot of Track Duration Across Artists')
plt.xticks(rotation=90)
plt.show()
```
